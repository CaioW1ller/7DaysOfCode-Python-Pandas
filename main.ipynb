### Anexar a categoria com base no nº(localizacao)


```python
def classificar_categoria(num):
    if 0 <= num <= 99:
        return("Generalidades. Ciência e conhecimento.")
    elif num >= 100 and num <= 199:
    	return("Filosofia e psicologia.")
    elif num >= 200 and num <= 299:
    	return("Religião.")
    elif num >= 300 and num <= 399:
    	return("Ciências sociais.")
    elif num >= 400 and num <= 499:
    	return("Classe vaga. Provisoriamente não ocupada.")
    elif num >= 500 and num <= 599:
    	return("Matemática e ciências naturais.")
    elif num >= 600 and num <= 699:
    	return("Ciências aplicadas.")
    elif num >= 700 and num <= 799:
    	return("Belas artes.")
    elif num >= 800 and num <= 899:
    	return("Linguagem. Língua. Linguística.")
    elif num >= 900 and num <= 999:
    	return("Geografia. Biografia. História.")
    else:
        return("Sem Categoria")

emprestimos_completo['classificacao_decimal_universal'] = emprestimos_completo['localizacao'].apply(classificar_categoria)
emprestimos_completo
```

### Função que será resposável por pegar o dataFrame e fazer o calculo da porcentagem dos cursos em base sua quantidade

```python
def percentual_ano(dataFrame):
    df = dataFrame.copy()
    df['Curso'] = df['Curso'].str.strip().str.upper()
    
    df = df.sort_values(by='Curso', ascending=True)
    
    df['Percentual'] = round((df['Quantidade'] / df['Quantidade'].sum()) * 100, 1)

    return df
```

### Importo para uma variável o arquivo 'previsao.txt', organizo deixando as letras maisculas e crio um DataFrame pro arquivo de modo a organizá-lo conforme o padrão dos outros anos

```python
def acentuacao(text):
    accents = {
        'administracao': 'administração',
        'arquitetura_e_urbanismo': 'arquitetura_e_urbanismo',
        'artes_cenicas': 'artes_cênicas',
        'bioinformatica': 'bioinformática',
        'bioquimica': 'bioquímica',
        'ciencia_e_engenharia_de_materiais': 'ciência_e_engenharia_de_materiais',
        'ciencia_tecnologia_e_inovacao': 'ciência_tecnologia_e_inovação',
        'ciencias_da_saude': 'ciências_da_saúde',
        'ciencias_odontologicas': 'ciências_odontológicas',
        'design': 'design',
        'engenharia_civil': 'engenharia_civil',
        'engenharia_software': 'engenharia_software',
        'filosofia': 'filosofia',
        'gestao_da_informacao_e_do_conhecimento': 'gestão_da_informação_e_do_conhecimento'
    }
    return accents.get(text, text)

dados = []

with open('previsao.txt', 'r') as arquivo:
    for linha in arquivo:
        curso, previsao = linha.strip().rsplit(' ', 1)
        curso_accentuated = acentuacao(curso)
        dados.append([curso_accentuated.replace('_', ' ').upper(), previsao])

p2022 = pd.DataFrame(dados, columns=['Curso', 'Quantidade'])
p2022 = p2022.drop(0)
p2022['Quantidade'] = p2022['Quantidade'].astype(int)
p2022 = p2022.reset_index(drop=True)

p2022
```

### Calcula as diferenças percentuais entre anos para cada curso e seleciona essas diferenças em um novo DataFrame chamado df_diferencas

```python
diferenca_percentual_ano['Diferença_2017_2018(%)'] = diferenca_percentual_ano['Percentual_2018(%)'] - diferenca_percentual_ano['Percentual_2017(%)']
diferenca_percentual_ano['Diferença_2018_2019(%)'] = diferenca_percentual_ano['Percentual_2019(%)'] - diferenca_percentual_ano['Percentual_2018(%)']
diferenca_percentual_ano['Diferença_2019_2022(%)'] = diferenca_percentual_ano['Percentual_2022(%)'] - diferenca_percentual_ano['Percentual_2019(%)']

df_diferencas = diferenca_percentual_ano[['Curso', 'Diferença_2017_2018(%)', 'Diferença_2018_2019(%)', 'Diferença_2019_2022(%)']]

df_diferencas
```

### Aplico uma coloração condicional às células de um DataFrame usando a biblioteca Pandas para estilização

```python
def cor(val):
    try:
        color = 'red' if val < 0 else 'green'
    except ValueError:
        color = 'black'
    return f'color: {color}'

df_styled = df_diferencas.style.map(cor, subset=['Diferença_2017_2018(%)', 'Diferença_2018_2019(%)', 'Diferença_2019_2022(%)'])

df_styled
```
